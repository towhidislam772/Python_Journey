```python
accuracy = 95

while True :
    if accuracy == 100:
        break
    accuracy+=1
```

```python
accuracy = 95
changed_or_not=True
prev=0
while True :
    if  changed_or_not == False and accuracy == 100:
        break
    if(prev>accuracy):
        changed_or_not=False
    accuracy+=1
    prev=accuracy
```
