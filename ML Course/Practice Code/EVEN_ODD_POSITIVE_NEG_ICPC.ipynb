```python
num_input=int(input())

inp=input()

numbers=inp.split()

even=0 
odd=0
positive=0
negative=0

for i in range(num_input):
    if int(numbers[i]) % 2==0:
        even+=1
    else:
        odd+=1

    if int(numbers[i])>0:
        positive+=1
    elif int(numbers[i])<0:
        negative+=1
    else:
        continue


print("Even:",even)
print("Odd:",odd)
print("Positive:",positive)
print("Negative:",negative)
```
