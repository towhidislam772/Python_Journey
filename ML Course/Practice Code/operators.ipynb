Arithmatic Operator

```python
x=10 ; y=3
sum = x+y
sub = x-y
mul = x*y
div = x/y
rem = x%y

div2 = round(x/y, 2)

print(sum, sub, mul, div, rem, div2)
```

Comparison Operator

```python
x=10 ; y=3

greater_than = x>y

greater_than_or_equal = (10 >=10)

print(greater_than, greater_than_or_equal)

less_than = x<y

less_than_or_equal = (10 <=10)

print(less_than, less_than_or_equal)

equal = x==y

print(equal)

not_equal =  x!=y

print(not_equal)
```
