### 5.4 Inheritance

##Single Inheritence

```python
#parent class / Parent class
class Phone:
  category="Electronics"

  #constructor
  def __init__(self,model,battery,camera,battery_percent=100): #__ duita hobe
    self.model=model
    self.battery=battery
    self.camera=camera
    self.battery_percent = battery_percent

  #methods
  def charge(self,hour):
    #self.charge +=hour
    print(f"charge completed by {hour}")

  def capture(self,total_photo):
    if(self.battery_percent)<=0:
      print("No charge")
    else:
      self.battery_percent-=total_photo
      print(f"Photo Captured in {self.model}")
```

```python
apple= Phone("Iphone 7",3000,40)
blueberry = Phone("V 17",4000,30)
motorola = Phone("M 17",3500,35)

print(motorola.battery_percent)
apple.capture(10)
print(apple.battery_percent)
```

```python
#child / derived class
class SmartPhone(Phone):
  def __init__(self,model,battery,camera,processor):
    super().__init__(model,battery,camera) #base class er init call hobe and value pass hobe
    self.processor=processor

  def charge(self,hour):
    print("first charging in process")
    super().charge(hour) # eta na korle parent er charge function ta call hobe na just child er ta print hobe


```

```python
pro= SmartPhone("X",500,100,"SnapDragon")
print(pro.model)

pro.charge(1)
```

##Multiple Inheritence

```python
#parent class / Parent class
class Phone:
  category="Electronics"

  #constructor
  def __init__(self,model,battery,camera,battery_percent=100): #__ duita hobe
    self.model=model
    self.battery=battery
    self.camera=camera
    self.battery_percent = battery_percent

  #methods
  def charge(self,hour):
    #self.charge +=hour
    print(f"charge completed by {hour}")

  def capture(self,total_photo):
    if(self.battery_percent)<=0:
      print("No charge")
    else:
      self.battery_percent-=total_photo
      print(f"Photo Captured in {self.model}")

#another class
class Cooling_mechanism:
  def __init__(self,cooling_method):
    self.cooling_method=cooling_method

  def cooling_on(self):
    print(f"the system is being cooled by {self.cooling_method}")
```

```python
#child / derived class
class SmartPhone(Phone):
  def __init__(self,model,battery,camera,processor):
    super().__init__(model,battery,camera) #base class er init call hobe and value pass hobe
    self.processor=processor

  def charge(self,hour):
    print("first charging in process")
    super().charge(hour) # eta na korle parent er charge function ta call hobe na just child er ta print hobe


```

```python
class SmartPhone_CoolingMode(SmartPhone,Cooling_mechanism):
  def __init__(self,model,battery,camera,processor,cooling_method):
    #super().__init__(model,camera,battery,processor)
    SmartPhone.__init__(self,model,camera,battery,processor)
    Cooling_mechanism.__init__(self,cooling_method)
```

```python
pro_cooling = SmartPhone_CoolingMode("Y",5000,35,"Octacore","Nitrogen")

print(pro_cooling.processor) # SmartPhone class theke
print(pro_cooling.battery) # Phone class theke
print(pro_cooling.cooling_method) # Cooling_Mechanism class theke
print(pro_cooling.cooling_on()) # Cooling_Mechanism class er method
print(pro_cooling.charge(1)) # SmartPhone er modified charge jeta Phone class theke inherited
```
