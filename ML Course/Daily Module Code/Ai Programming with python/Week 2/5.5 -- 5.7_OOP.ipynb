###5.5 Polymorphism

```python
class Camera:
  def __init__(self,name):
    self.name=name

  def capture(self):
    print("captured")
```

```python
class Smart(Camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution=resolution

  def capture(self): #overriding the method
    print("Photo is captured by a phone")

class DSLR(Camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution=resolution

  def capture(self): #overriding the method
    print("Photo is captured by a DSLR")


class Drone(Camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution=resolution

  def capture(self): #overriding the method
    print("Photo is captured by a Drone")

phone=Smart("Phone",30)
dslr=DSLR("Dslr",100)
drone=Drone("Drone",150)

phone.capture()
dslr.capture()
drone.capture()
```

###5.6 Encapsulation

```python
class Mobile:
  def __init__(self,name,model,imei):
    self.name=name
    self.model=model
    self.imei=imei

  def charge(self):
    print("Phone is charging")

iphone=Mobile("Phone","17","1xFac")

print(iphone.name)
print(iphone.model)
print(iphone.imei)
```

```python
iphone.name="Iphone"
print(iphone.name)

iphone.imei="shODn"
print(iphone.imei)
```

```python
class Mobile:
  def __init__(self,name,model,imei):
    self.__name=name
    self.__model=model
    self.__imei=imei # __ double for making it private

  def charge(self):
    print("Phone is charging")

  def imei_getter(self): #only method to get imei number private
    return self.__imei

  def model_getter(self): #only method to get model  private
    return self.__model

  def name_getter(self): #only method to get name  private
    return self.__name

  def name_setter(self,name): #only way to change name
    self.__name=name

iphone=Mobile("Phone","17","1xFac")

print(iphone.name_getter())
print(iphone.model_getter())
print(iphone.imei_getter())

iphone.imei="sadfasd" #wont change private attribute
print(iphone.imei_getter())

iphone.name_setter("Phitron") # will change the name
print(iphone.name_getter())
```

```python
from abc import ABC,abstractmethod
class Telephone(ABC):

  @abstractmethod
  def make_call(self):
    print("making call")


class Sphone(Telephone):
  def make_call(self): #i have to inherit abstract method to create instance
    print("making call from sphone")

class Iphone(Telephone):
  def make_call(self):
    print("making call from Iphone")
```

```python
ip=Iphone()
ip.make_call()
```
