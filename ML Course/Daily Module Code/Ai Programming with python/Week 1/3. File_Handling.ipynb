####3.4 File Pointer

```python
with open("./sample_data/sample.txt","r") as file:
  print(file.tell()) # currently cursor koi ase
```

```python
with open("./sample_data/sample.txt","r") as file:
  print(file.tell()) # currently cursor koi ase
  print(file.read()) #sob pore last e jabe
  print(file.tell())
```

```python
with open("./sample_data/sample.txt","r") as file:
  print(file.tell()) # currently cursor koi ase
  print(file.read())
  print(file.tell())
  ## last index theke pora shuru korsi
  print(file.read())
```

```python
with open("./sample_data/sample.txt","r") as file:
  print(file.tell())
  print(file.read(5))
  print(file.tell())
```

```python
with open("./sample_data/sample.txt","r") as file:
  print(file.tell())
  print(file.read(5))
  print(file.tell())

  print(file.read()) #by default -1 thake mane sob

  print(file.tell())
```

```python
#seek

with open("./sample_data/sample.txt","r") as file:
  print(file.tell())
  file.seek(5)
  print(file.tell())
```

```python
with open("./sample_data/sample.txt","r") as file:
  print(file.tell())

  print(file.read(5))

  print(file.tell())

  file.seek(0) #exact position e niye asha
  print(file.tell())
  print(file.read(5))
```
