Practice 1 :
Get the text from sample.txt , and count the following

Number of lines

Number of words

Number of characters

and save it to the counter_of_string.txt file

```python
from functools import reduce
with open ("./sample_data/sample.txt", "r") as file:
    string=file.readlines()

    total_lines=len(string)

    print(total_lines)

    number_of_words = list(map(lambda x: len(x.split()),string))
    print(number_of_words)

    total_number_of_words= reduce(lambda x,y :x+y, number_of_words)

    print(total_number_of_words)

    string = list(map(str.strip, string))

    string= list(map(lambda x: x.replace(" ",""),string))

    number_of_character = list(map(lambda x : len(x),string))

    print(number_of_character)

    total_number_of_characters = reduce(lambda x,y : x+y ,number_of_character)

    print(total_number_of_characters)

    print(string)

    with open("./sample_data/counter_of_string.txt","w") as file:
      file.write(f"total line: {total_lines}\ntotal number of words:{total_number_of_words}\ntotal number of char: {total_number_of_characters}")


```

Create a text file named emails.txt containing multiple email addresses, some in uppercase and some repeated.
Write a Python program to clean the data by converting all emails to lowercase, removing duplicates, sorting them alphabetically, and saving the cleaned list to a new file named unique_emails.txt.


```python
with open("./sample_data/emails.txt","r") as file:
  content = file.readlines()
  print(content)

  content=list(map(str.strip,content))
  print(content)

  content=list(map(str.lower,content))
  print(content)

  content = set(content)
  print(content)

  content = sorted(content)
  print(content)

  with open("./sample_data/unique_emails.txt","w") as file:
    file.write("\n".join(content))

```

Create a text file named names.txt containing a list of names.
Some names may have extra spaces, inconsistent capitalization (like “aLiCe” or “ bob ”), or appear more than once.
Write a Python program to clean the data by:
1.  Removing extra spaces.
2.  Converting all names so that only the first letter is capitalized.
3.  Removing duplicate names.
4. Sorting the names alphabetically.
Finally, save the cleaned list into a new file named clean_names.txt.


```python
with open("./sample_data/names.txt","r") as file:
  content = file.readlines()
  print(content)

  content=list(map(str.strip,content))
  print(content)

  content=list(map(str.capitalize,content))
  print(content)

  content = set(content)
  print(content)

  content = sorted(content)
  print(content)

  with open("./sample_data/clean_names.txt","w") as file:
    file.write("\n".join(content))
```
