```python
def adder(a,b):
    print(f"{a},{b}")

adder(a=10,b=20)
```

```python
def adder(a,b,c):
    print(f"{a},{b},{c}")

adder(a=10,c=30,b=20)
```

```python
def adder(**kwargs):
    print(kwargs)

adder(a=10,c=30,b=20, n_estimator = 50)
```

```python
def adder(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} : {value}")

adder(a=10,c=30,b=20, n_estimator = 50)
```
